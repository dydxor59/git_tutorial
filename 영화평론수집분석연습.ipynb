### 영화평론수집 분석 연습

```python
import requests,re
import sqlite3
from bs4 import BeautifulSoup
dbname='namovie.db'
```

```python
def mycrawl(soup):
    tgt=soup.select('.list_netizen tbody > tr')

    data=[]
    with sqlite3.connect(dbname) as conn:
        cur=conn.cursor()
        sql='insert into movie (IDX,TITLE,CONTENT,POINT) values(?,?,?,?)'
        for t in tgt:

            idx=t.select('td.num')[0].get_text()
            print(idx)

            # 중복제거
            sql_sel='select IDX from movie where IDX='+idx
            resultid=cur.execute(sql_sel).fetchall()

            if resultid:
                print('중복에러'+idx,end="\n")

            else:
                idx=int(idx)

                point=int(t.select('td.point')[0].get_text())
                print(point)

                title=t.select('a.movie')[0].get_text()
                print(title)

                content=t.select('td.title')[0].get_text()
                c=content.replace(title,'').replace('신고','').strip()
                print(c)

                d=[idx,title,c , point]
                print(d)

                cur.execute(sql,d)

        conn.commit()
        cur.close()
    
```

```python
for i in range(10,20):
    addr='https://movie.naver.com/movie/point/af/list.nhn?&page='
    addr+=str(i)
    result=requests.get(addr)
    src=result.content
    # print(result.content)

    soup=BeautifulSoup(src,'html.parser')
    print(addr,end="\n")
    mycrawl(soup)
```

```python
with sqlite3.connect(dbname) as conn:
    cur=conn.cursor()
    sql='create table movie(IDX integer primary key, TITLE varchar(100), CONTENT varchar(200), POINT integer)'
    cur.execute(sql)
    conn.commit()
    cur.close()
```
